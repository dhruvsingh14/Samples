```python
df = df.fillna(0)
```

```python
df = df.rename(columns = {' Contract Amount ': 'Contract Amount'})
```

```python
df['Contract Amount'] = df['Contract Amount'].str.replace('$','')
```

```python
df['Contract Amount']
```

```python
df['Contract Amount'] = df['Contract Amount'].str.replace(',','')
```

```python
df['Contract Amount']
```

```python
df['Contract Amount'] = df['Contract Amount'].str.replace(' ','')
```

```python
df['Contract Amount'] = df['Contract Amount'].str.replace('-','')
```

```python
df['Contract Amount']
```

```python
df['Contract Amount'] = pd.to_numeric(df['Contract Amount'])
```
