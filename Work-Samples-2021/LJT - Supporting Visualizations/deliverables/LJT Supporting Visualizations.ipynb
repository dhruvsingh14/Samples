#### Funding

```python
Funding = Funding.set_index('Program')
```

```python
Funding.head()
```

```python
plt.style.use('ggplot')

x = Funding.index.tolist()

x_pos = [i for i, _ in enumerate(x)]

plt.bar(x_pos, Funding['FY21 Annual Budget'], color='seagreen')
plt.xlabel("Program")
plt.ylabel("Budget ($ millions)")
plt.title("FY 2021 Budget")

plt.xticks(x_pos, x)

plt.savefig('graphs/1_Funding.png', dpi=300, bbox_inches='tight')
plt.show()
```
