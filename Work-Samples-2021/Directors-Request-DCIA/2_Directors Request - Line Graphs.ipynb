```python
# libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('cleaned_data.csv')
```

```python
df.head()
```

#### Subsetting 2: Line Graphs

```python
df1 = df[['State Id', 'ServiceProviderName', 'CourseName', 'ExitDate', 'Year']]
```

```python
df1 = df1.sort_values(by=['Year', 'State Id', 'ExitDate'], ascending = (True, True, False))
```

```python
df1.shape
```

```python
df1 = df1.drop_duplicates(subset = ['State Id'], keep='first')
```

```python
df1.head()
```

```python
df1.shape
```

#### Line Graph 1: Total Number by Service Provider

```python
df2 = df1[['Year', 'ServiceProviderName', 'State Id']].groupby(['Year', 'ServiceProviderName']).count()
```

```python
df2
```

```python
df2 = df2.reset_index()
df2 = df2.rename(columns = {"State Id": "Number Employed"})
```

```python
df2.head()
```

```python
df2_long = df2.pivot_table(index='Year', columns='ServiceProviderName', values='Number Employed').reset_index()
```

```python
df2_long = df2_long.fillna(0)
```

```python
df2_long = df2_long.set_index('Year')
```

```python
df2_long
```

```python
# df2_long.to_csv('service_provider.csv')
```

```python
df2_long = df2_long[['The Training Zone of the DMV', 'DC Department of Employment Services', 'CC Prep - MLK', 'GRID Alternatives Mid-Atlantic', 'University of the District of Columbia (UDC)']]
```

```python
df2_long = df2_long.rename(columns = {'The Training Zone of the DMV': 'DMV', 
                                      'DC Department of Employment Services': 'DOES', 
                                      'CC Prep - MLK': 'MLK', 
                                      'GRID Alternatives Mid-Atlantic': 'GRID', 
                                      'University of the District of Columbia (UDC)': 'UDC'})
```

```python
## comparing service providers
df2_long.plot(kind='line')
plt.title('Total Employment by Service Provider')
plt.ylabel('Number Employed')
plt.xlabel('Year')
plt.xticks([2018, 2019, 2020, 2021])
plt.savefig('graphs/service_provider.png', dpi=300, bbox_inches='tight')
plt.show()
```

#### Line Graph 2: Total Number by Course Name

```python
df3 = df1[['Year', 'CourseName', 'State Id']].groupby(['Year', 'CourseName']).count()
```

```python
df3
```

```python
df3 = df3.reset_index()
df3 = df3.rename(columns = {"State Id": "Number Employed"})
```

```python
df3_long = df3.pivot_table(index='Year', columns='CourseName', values='Number Employed').reset_index()
```

```python
df3_long = df3_long.fillna(0)
```

```python
df3_long = df3_long.set_index('Year')
```

```python
df3_long
```

```python
# df3_long.to_csv('course_title.csv')
```

```python
df3_long = df3_long[['Services to Individuals Seeking Employment', 'Flagger/OSHA 10 Safety Training', 'Help Desk IT Training', 'Solar Works', 'Quick Path to Energy', 'OSHA 30']]
```

```python
df3_long = df3_long.rename(columns = {'Services to Individuals Seeking Employment': 'Serv Indiv.', 
                                      'Flagger/OSHA 10 Safety Training': 'OSHA 10', 
                                      'Help Desk IT Training': 'HlpDsk IT', 
                                      'Solar Works': 'Solar', 
                                      'Quick Path to Energy': 'Qck Pth', 
                                      'OSHA 30': 'OSHA 30'})
```

```python
## comparing course titles
df3_long.plot(kind='line')
plt.title('Total Employment by Course Title')
plt.ylabel('Number Employed')
plt.xlabel('Year')
plt.xticks([2018, 2019, 2020, 2021])
plt.savefig('graphs/course_title.png', dpi=300, bbox_inches='tight')
plt.show()
```
