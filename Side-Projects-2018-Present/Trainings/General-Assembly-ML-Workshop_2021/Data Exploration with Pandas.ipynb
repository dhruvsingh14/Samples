What was the survival rate for adult men (age>=18)

```python
(train.Age>=18) & (train.Sex=='male') # boolean for male over 18, useful for indexing
```

What was the survival rate for women and children?

```python
train[(train.Sex=='female')|(train.Age<18)] # using symbols as opposed to english 'and', 'or'
```

Use groupby to compare the survival rates of men and women

```python
train.groupby(['Sex', 'Embarked'])[['Age', 'Fare', 'Survived']].mean() # can group by category
```

What was the average age of the survivors?

```python
train[train.Survived == 1].Age.mean()
```

What was the combined survival rate of both children (age less than 18) and seniors (age greater than 60)?

```python
train.Survived[(train.Age<18) | (train.Age>60)].mean() # averaging binary provides a rate
```

Group by Pclass and investigate average survival rate, age and fare

```python
train.groupby(['Sex', 'Pclass'])[['Survived', 'Age', 'Fare']].mean() # can group by category
# sex seems to impact survival more than class

# double bracket here is not list of lists, is instead indexing using a list
```
